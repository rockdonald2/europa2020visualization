```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
df = pd.read_json('../data/europa-data.json', orient='index')
```

```python
fig, axis = plt.subplots(ncols=1, nrows=1, figsize=(16, 4))

df.loc['BE'][['ge_2015', 'ge_2016', 'ge_2017', 'ge_2018']].plot(kind='line', grid=True, ax=axis)
target = df.loc['BE']['ge_t']
temp = pd.Series(target, index=['ge_2015', 'ge_2016', 'ge_2017', 'ge_2018'])
temp.plot(kind='line', ax=axis, grid=True)
```
